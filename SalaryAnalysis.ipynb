# Data Preparation

## Write a python program to load data into pandas DataFrame.

```python
import pandas as pd

#reading the "DataScienceSalaries.csv" into pandas dataframe and assigning variable name data
data=pd.read_csv("DataScienceSalaries.csv")
data
```

## Write a python program to remove unnecessary columns i.e., salary and salary currency.

```python
#dropping the columns salary and salary_currency
data=data.drop(columns=['salary','salary_currency'])
data
```

## Write a python program to remove the NaN missing values from updated dataframe. 

```python
#dropping the rows which has empty cells
data=data.dropna()
data
```

# Data Analysis

## 	Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable

```python
from scipy.stats import skew, kurtosis

#finding sum, mean, standard deviation, skewness and kurtosis
Answer = {
    'Sum': data['salary_in_usd'].sum(),
    'Mean': data['salary_in_usd'].mean(),
    'Standard Deviation': data['salary_in_usd'].std(),
    'Skewness': skew(data['salary_in_usd']),
    'Kurtosis': kurtosis(data['salary_in_usd'])
}
print(Answer)
```

# Data Exploration

## Write a python program to find out top 15 jobs. Make a bar graph of sales as well.         

```python
#finding the top 15 job_titles on basis of number of jobs available
top_fifteen_jobs = data['job_title'].value_counts().head(15)
top_fifteen_jobs
```

```python
#showing number of jobs and job titles in bar graph
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top_fifteen_jobs.plot(kind='bar', color='lightblue')
plt.title('Top 15 Jobs')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

##	Which job has the highest salaries? Illustrate with bar graph.	

```python
#showing the job title with highest salaries in bar graph
mean_salary_by_job = data.groupby('job_title')['salary_in_usd'].mean()

#storing data of top 40 highest mean salaries of job titles
highest_40_mean_salary=mean_salary_by_job.sort_values(ascending = False).head(40)

#getting highest salary and it's title
highest_mean_salary_job = highest_40_mean_salary.idxmax()
highest_mean_salary = mean_salary_by_job.max()

#forming bar graph
plt.figure(figsize=(15, 8))
highest_40_mean_salary.plot(kind='bar', color='lightblue')
plt.title('Top 40 Job Title by Average Salary')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=90, ha='right')
plt.axhline(y=highest_mean_salary, color='r', linestyle='--', label=f'Highest Mean Salary: {highest_mean_salary:.2f} USD ({highest_mean_salary_job})')
plt.legend()

plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

## 	Write a python program to find out salaries based on experience level. Illustrate it through bar graph.	

```python
#finding average salary by experience level
mean_salary_by_experience = data.groupby('experience_level')['salary_in_usd'].mean()

plt.figure(figsize=(6, 8))
mean_salary_by_experience.plot(kind='bar', color='lightblue')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=90, ha='right')
plt.legend()
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

## Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph. 

```python
import numpy as np

#histrogram showing number of jobs and salary 
salary=data['salary_in_usd']
b=int((data['salary_in_usd'].max()-data['salary_in_usd'].min())/len(data))

#creating histrogram
plt.figure(figsize=(18, 7))
plt.hist(salary, bins=b, color='skyblue',edgecolor='white')
plt.grid(axis='y')

#setting x label 
xticks_locations = np.linspace(salary.min(), salary.max(), num=b)
plt.xticks(xticks_locations,rotation=90,fontsize=8)

plt.xlabel('Salary in USD')
plt.ylabel('Number of jobs in Data Science')
plt.title('Salary in Data Science')

plt.show()
```

```python
#box plot using company size
size=data['company_size'].value_counts()
size.plot(kind='box',title='Box chart by Company Size')
plt.show()
```
